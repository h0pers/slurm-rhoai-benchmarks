# Scenario 6: Job Queuing and Prioritization

| | |
|---|---|
| **Workload class** | Job queuing and prioritization |
| **Matrix status** | Partially Ready (no true backfill, no job arrays) |
| **Shape** | Low-priority holds 6 GPUs, high-priority needs 4 GPUs |

**What this tests:** Priority-based preemption. When a high-priority job needs GPUs that a low-priority job is using, Kueue must evict the low-priority workload.

**How it works:** A low-priority job grabs 6 GPUs. After 60 seconds, a high-priority job requesting 4 GPUs is submitted (6 + 4 > 8 quota). Kueue evicts the low-priority workload, the high-priority job runs, and the low-priority workload is requeued afterward.

**Success criteria:**
- Low-priority workload is evicted
- High-priority job completes
- Low-priority workload is requeued after eviction
- Preemption latency is recorded

**Slurm equivalent:** Slurm supports preemption with configurable policies (cancel, checkpoint, requeue). Kueue provides the same - higher-priority work displaces lower-priority work. Note: Kueue does not auto-checkpoint before preemption.
