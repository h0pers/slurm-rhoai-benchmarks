# Scenario 7: HPC Mathematical Algorithms (Not Benchmarked)

| | |
|---|---|
| **Workload class** | HPC mathematical algorithms |
| **Matrix status** | Not Ready |
| **Shape** | Not benchmarked |

## What kind of work falls into this category?

Some scientific problems - weather simulations, drug discovery, physics
modeling - need many GPUs to work together at the same time, exchanging
data back and forth thousands of times per second. Think of it like a
team of calculators that must constantly share their intermediate answers
to solve one giant equation together. This is called "tightly-coupled"
computing.

For this to work well, the GPUs need a very fast, direct connection
between them - much faster than regular network cables can provide.

## Why we did not benchmark this

The networking technology these workloads need is not ready for
production use yet:

| What is missing | Current state in RHOAI 3.4 |
|---|---|
| **RoCE / GPUDirect RDMA** - a way for GPUs to talk directly to each other over the network, skipping the CPU entirely | Available as a preview, but not officially supported for production |
| **Network topology awareness in Kueue** - the scheduler knowing which GPUs are physically close to each other on the network | Not available |
| **Mixed job types** - running jobs that need both GPU and CPU-only machines together | Not available |

Without these features, running a benchmark would mostly measure how
slow regular networking is - not how well the platform handles the
actual workload. That would give misleading results, so we chose to
skip this for now.

## What we plan to test once the technology is ready

When RoCE (the fast GPU-to-GPU networking) becomes officially supported,
we would run well-known scientific computing benchmarks:

| Benchmark | What it tells us (in plain terms) |
|---|---|
| [HPL](https://catalog.ngc.nvidia.com/orgs/nvidia/containers/hpc-benchmarks) (High-Performance Linpack) | How fast can the system solve a massive math problem? This is the same test used to rank the world's fastest supercomputers. |
| [HPCG](https://catalog.ngc.nvidia.com/orgs/nvidia/containers/hpc-benchmarks) (High-Performance Conjugate Gradients) | How well does the system handle problems that jump around in memory rather than crunching through data in order? This stresses the network between GPUs. |
| [PyHPC-Benchmarks](https://github.com/dionhaefner/pyhpc-benchmarks) | Scientific equations implemented in Python - tests the kind of code researchers actually write. |

We would run these benchmarks across 2 or more machines with the fast
GPU networking turned on, and compare the results against the same test
on Slurm (the industry-standard system for running this type of work).

The key numbers we would look at:
- **GFLOPS** - raw computing speed (billions of math operations per second)
- **Inter-node bandwidth** - how fast data moves between machines
- **GPU-to-GPU latency** - how long it takes for one GPU to send a message to another

**Bottom line:** This category of work is on hold until the fast
networking features are production-ready in RHOAI. Once they are, we
will revisit it.
