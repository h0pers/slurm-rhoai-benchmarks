# Scenario 5: Gang Scheduling

| | |
|---|---|
| **Workload class** | Gang-scheduled jobs |
| **Matrix status** | Ready |
| **Shape** | Blocker holds 4 GPUs, measured job needs 6 GPUs (2 nodes x 3) |

**What this tests:** Kueue's all-or-nothing admission. A job that needs more GPUs than are available must wait completely (zero pods created) until enough GPUs are free, then start all workers together.

**How it works:** A blocker job grabs 4 of 8 quota GPUs for 3 minutes. The measured job needs 6 GPUs (4 + 6 > 8), so Kueue must hold it fully pending. When the blocker finishes, Kueue admits the measured job and all pods start together.

**Success criteria:**
- Zero pods created for the measured job while the blocker runs
- Measured job admitted only after blocker completes
- Both measured pods start within a small window (gang start spread)

**Slurm equivalent:** Slurm gang scheduling waits for all resources before starting any task. This confirms Kueue provides the same guarantee.
