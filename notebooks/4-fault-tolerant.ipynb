# Scenario 4: Fault-Tolerant Training with Checkpointing

| | |
|---|---|
| **Workload class** | Fault-tolerant training with checkpointing |
| **Matrix status** | Partially Ready (JIT checkpointing is Early Access) |
| **Shape** | 2 nodes, 2 GPUs each, 600 steps, NFS checkpoint + worker kill |

**What this tests:** Whether a training job can survive a worker pod being killed, restart, and resume from the last checkpoint instead of starting over.

**How it works:** The script saves checkpoints to a shared NFS PVC. After 60 seconds, the harness kills one worker pod. The job controller restarts, and the trainer resumes from the saved checkpoint.

**Success criteria:**
- Job completes after the pod kill
- `resumed_from_step > 0` in logs (resumed from checkpoint, not step 0)
- Recovery time (kill to pods running again) is recorded

**Slurm equivalent:** Checkpoint/restart via DMTCP or user scripts. RHOAI's JIT checkpointing achieves the same with zero code changes.
